```python
import ee
import geemap

ee.Initialize(project="celtic-spider-499508-n9")

# 先用黄河下游/入海前附近点。
# 如果结果面积不对，后面只需要微调这个点。
outlet = ee.Geometry.Point([118.55, 37.50])

levels = [5, 6, 7, 8, 9]

for lev in levels:
    fc = ee.FeatureCollection(f"WWF/HydroSHEDS/v1/Basins/hybas_{lev}")

    outlet_feat = ee.Feature(fc.filterBounds(outlet).first())
    props = outlet_feat.toDictionary().getInfo()

    main_bas = props["MAIN_BAS"]
    hybas_id = props["HYBAS_ID"]

    basin = fc.filter(ee.Filter.eq("MAIN_BAS", main_bas))
    area_km2 = basin.geometry().area(1).divide(1e6).getInfo()

    print("================================")
    print("Level:", lev)
    print("Outlet HYBAS_ID:", hybas_id)
    print("MAIN_BAS:", main_bas)
    print("Area km2:", area_km2)
    print("Outlet feature properties:")
    print(props)
```
